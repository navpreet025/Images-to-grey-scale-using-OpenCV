```python
import os
import shutil
import cv2
from bing_image_downloader import downloader

# Step 1: Create the input_images folder
input_folder = 'input_images'
os.makedirs(input_folder, exist_ok=True)

# Step 2: Download 100 images using Better Bing Image Downloader
search_keyword = "flowers"  # You can change this to anything
downloader.download(search_keyword, limit=100, output_dir=input_folder, adult_filter_off=True)
# ault filtre 0ff= allow all types of files no filtre
#output_dir=input_folder?= because we want to the subfolder to be created inside input_images and thyen we move its content up for further processing
# Step 3: Move images from subfolder (e.g., input_images/flowers) to input_images/
subfolder_path = os.path.join(input_folder, search_keyword) #build full path to subfolder where bing saves the images(e.g., input_images/flowers)

for file in os.listdir(subfolder_path):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        shutil.move(os.path.join(subfolder_path, file), os.path.join(input_folder, file))

os.rmdir(subfolder_path)  # Remove the now-empty subfolder

print(f"✅ Downloaded and moved 100 '{search_keyword}' images to '{input_folder}'")

# Step 4: Convert images to grayscale using OpenCV
output_folder = 'output_images'
os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_folder, file) #build full path of image
        img = cv2.imread(img_path)

        if img is None:
            print(f"⚠️ Skipped unreadable image: {file}")
            continue

        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        output_path = os.path.join(output_folder, file)
        cv2.imwrite(output_path, gray)

print(f"✅ All images converted to grayscale and saved in '{output_folder}'")
```
